```python
def a_ve_b_bul(x,y):
    a=sum((x-x.mean())*(y-y.mean()))/sum((x-x.mean())**2)
    b=y.mean()-a*x.mean()
    return a, b
a, b = a_ve_b_bul(veri["yuksekogretim_mezuniyet_orani"], veri["is_gucune_katilim_orani"])
print(a,b)
```
